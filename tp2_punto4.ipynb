## Punto 4
Se presenta un sistema con realimentación negativa, con un disturbio negativo introducido entre el controlador y la planta

```
                                        │ D(s)
                                        │
                  ┌────────────┐        ▼ -     ┌────────────┐
R(s)  ┌───┐ E(s)  │            │ U(s) ┌───┐     │            │    Y(s)
─────►│   ├──────►│   G_c(s)   ├─────►│   ├────►│   G_p(s)   ├───┬─────►
      └───┘       │            │      └───┘     │            │   │
        ▲ -       └────────────┘                └────────────┘   │
        │                                                        │
        │                         ┌────────────┐                 │
        │                         │            │                 │
        └─────────────────────────┤   G_s(s)   │◄────────────────┘
                                  │            │
                                  └────────────┘
```
Por superposición, la función $Y(s)$ será la suma de los efectos causados por la perturbación y la excitación, por lo que para obtener su valor se propone dejar en cero alguna de las entradas, y calcular los efectos producidos por la otra.

Así, si comenzamos anulando la perturbación, el sistema tendrá la siguiente forma:

```
                  ┌────────────┐        ┌────────────┐
R(s)  ┌───┐ E(s)  │            │ U(s)   │            │    Y(s)
─────►│   ├──────►│   G_c(s)   ├───────►│   G_p(s)   ├───┬─────►
      └───┘       │            │        │            │   │
        ▲ -       └────────────┘        └────────────┘   │
        │                                                │
        │                  ┌────────────┐                │
        │                  │            │                │
        └──────────────────┤   G_s(s)   │◄───────────────┘
                           │            │
                           └────────────┘
```

Donde la función de transferencia $\frac{Y(s)}{R(s)}$ no será más que la función de transferencia de lazo cerrado con $G_c(s)\cdot G_p(s)$ en el lazo directo, y $G_s(s)$ en el lazo de realimentación.

$$\frac{Y(s)}{R(s)}=\frac{G_c(s) G_p(s)}{1+G_c(s) G_p(s) G_s(s)}$$

Por otro lado, anulando la excitación, obtenemos el siguiente diagrama:

```
                                  │ D(s)
                                  │
            ┌────────────┐        ▼ -     ┌────────────┐
┌───┐ E(s)  │            │ U(s) ┌───┐     │            │    Y(s)
│   ├──────►│   G_c(s)   ├─────►│   ├────►│   G_p(s)   ├───┬─────►
└─┬─┘       │            │      └───┘     │            │   │
  │ -       └────────────┘                └────────────┘   │
  │                                                        │
  │                         ┌────────────┐                 │
  │                         │            │                 │
  └─────────────────────────┤   G_s(s)   │◄────────────────┘
                            │            │
                            └────────────┘
```

Reacomodando los bloques:

```
                         ┌────────────┐
 -D(s) ┌───┐             │            │               Y(s)
 ─────►│   ├────────────►│   G_p(s)   ├──────────────┬─────►
       └───┘             │            │              │
         ▲ -             └────────────┘              │
         │                                           │
         │      ┌────────────┐      ┌────────────┐   │
         │ U(s) │            │ E(s) │            │   │
         └──────┤   G_c(s)   │◄─────┤   G_s(s)   │◄──┘
                │            │      │            │
                └────────────┘      └────────────┘
```

Tenemos que la función de transferencia con respecto a la perturbación toma nuevamente la forma de un sistema a lazo cerrado, donde ahora el lazo directo es $G_p(s)$ y el de realimentación $G_c(s)\cdot G_s(s)$. Es de notar qué dado al signo negativo en la entrada, la función de transferencia tendrá ganancia negativa.

$$\frac{Y(s)}{-D(s)}=\frac{G_p(s)}{1+G_p(s) G_c(s) G_s(s)}$$

$$\therefore$$

$$\frac{Y(s)}{D(s)}=-\frac{G_p(s)}{1+G_p(s) G_c(s) G_s(s)}$$

Finalmente $Y(s)$ para ambas entradas será:

$$Y(s)=R(s)\frac{G_c(s) G_p(s)}{1+G_c(s) G_p(s) G_s(s)}-D(s)\frac{G_p(s)}{1+G_p(s) G_c(s) G_s(s)}$$

Nótese que ambas expresiones comparten el denominador, por lo que si ninguno de los numeradores posee ceros distintos que anulen a los polos del denominador, las respuestas para cada entrada deberían presentar formas similares
